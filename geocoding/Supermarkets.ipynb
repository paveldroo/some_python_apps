```python
geolocator = geopy.Nominatim()
```

```python
address_list = []
index = 0
for i in df0.index:
    address_list.append(i + ' ' + df0['City'][index])
    index += 1
print(address_list)

address_latitudes = []
address_longitudes = []
for i in address_list:
    location = geolocator.geocode(i)
    try:
        address_latitudes.append(location.latitude)
        address_longitudes.append(location.longitude)
    except AttributeError:
        address_latitudes.append('Not Found')
        address_longitudes.append('Not Found')
print(address_latitudes)
print(address_longitudes)

```
